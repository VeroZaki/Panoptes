```python
# Program to check if a number is prime or not

def Get_prime(num):
  num = int(num)
  # To take input from the user
  #num = int(input("Enter a number: "))

  # define a flag variable
  flag = False

 # If given number is greater than 1
  if num > 1:
  
      # Iterate from 2 to n / 2
      for i in range(2, int(num/2)+1):
  
          # If num is divisible by any number between
          # 2 and n / 2, it is not prime
          if (num % i) == 0:
              flag =  False
              break
      else:
          flag =  True
  
  else:
      flag =  False


  # check if flag is True
  if flag == False:
      return False
  else:
      return True
```

```python
def label_video_names(in_dir):
    
    # list containing video names
    names = []
    # list containin video labels [1, 0] if it has violence and [0, 1] if not
    labels = []
    
    
    for current_dir, dir_names,file_names in os.walk(in_dir):
        
        for file_name in file_names:
            
            if file_name[0:1] == 'V' or file_name[0:2] == 'fi' or file_name[0:3] == 'new':   #fi
            #if file_name[0:1] == 'V':   #fi
                #print(file_name[3:4])
                labels.append([1,0])
                names.append(file_name)
            elif file_name[0:2] == 'NV' or file_name[0:2] == 'no':   #no
            #elif file_name[0:2] == 'NV':   #no
                labels.append([0,1])
                names.append(file_name)
                     
            
    c = list(zip(names,labels))
    # Suffle the data (names and labels)
    shuffle(c)
    
    names, labels = zip(*c)  # 

    #d=zip(*p)
    #list(d)
    #[(1, 4), (2, 5), (3, 6)]   Kida kol wa7da 2osadha al label beta3ha 
            
    return names, labels
```

## Generator that process one video through VGG16 each function call

```python
def proces_transfer(vid_names, in_dir, labels):
    
    count = 0
    
    tam = len(vid_names)
    
    # Pre-allocate input-batch-array for images.
    #shape = (_images_per_file,) + img_size_touple + (3,)   # Mabisht3'alish
    
    while count<tam:
        
        video_name = vid_names[count]
        
        #image_batch = np.zeros(shape=shape, dtype=np.float16)  # Mabisht3'alish 
    
        image_batch , cou = get_frames(in_dir, video_name)
        #print("frames: " , _images_per_file)  #cou
        #_images_per_file = cou
        
         # Note that we use 16-bit floating-points to save memory.
        shape = (_images_per_file, transfer_values_size)   # _images_per_file   # cou
        transfer_values = np.zeros(shape=shape, dtype=np.float16)
        
        #transfer_values = \
         #   image_model_transfer.predict(image_batch)
        transfer_values = image_model_transfer.predict(image_batch)
        #print("transfer_values : " , transfer_values)
        
        labels1 = labels[count]
        
        #aux = np.ones([41,2])
        aux = np.ones([_images_per_file ,2])  #_images_per_file   # cou
        
        labelss = labels1*aux
        
        yield transfer_values, labelss   # Return generator like list so we will read from it one by one (next())
        
        count+=1
```

## Define LSTM architecture

```python
chunk_size = 4096    # transfer_values_size -> aly hia a5ir layer fc2 al dense
n_chunks = 40        # al frames -> 40
rnn_size = 512       # Filter size

model = Sequential()
model.add(LSTM(rnn_size, input_shape=(n_chunks, chunk_size)))
model.add(Dropout(0.5))
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(50))
model.add(Activation('sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```
